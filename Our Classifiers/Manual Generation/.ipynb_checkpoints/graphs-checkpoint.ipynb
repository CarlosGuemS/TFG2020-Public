```python
import matplotlib.pyplot as plt
import numpy as np
```

## Code to generate the graphs

```python
# Method labels
labels = ["Base", "Base + TD", "Base + EMI", "Base + TD\n+ EMI"]
x = np.arange(len(labels))
```

```python
width = 0.15  # the width of the bars
sep_constant = 0.02

fig, ax = plt.subplots(figsize = (12, 8))


#Data
rectsOA = ax.bar(x - 3*width/2-1.5*sep_constant, data_OA, width, label='OA')
rectsOAE = ax.bar(x - width/2-0.5*sep_constant, data_OAE, width, label='OA-E')
rectsIWA = ax.bar(x + width/2+0.5*sep_constant, data_IWA, width, label='IWA')
rectsDLR = ax.bar(x + 3*width/2+1.5*sep_constant, data_DLR, width, label='DLR')

#The graph itself
ax.set_ylabel('F-Score (%)', size="x-large")
ax.set_xlabel('Set of features', size="x-large")
ax.set_xticks(x)
ax.set_yticks(np.arange(0,1.1,0.1))
ax.set_xticklabels(labels, size="x-large")
ax.tick_params(labelsize = "large")
ax.legend(loc = 'lower right', prop={'size': "large"})
legend = ax.get_legend()
legend.set_title("Datasets", prop={'size': "x-large"})
ax.grid(True, axis='y', alpha=0.5)

fig.tight_layout()

plt.savefig(file_title, format='pdf')
plt.show()
```
